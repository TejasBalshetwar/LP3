```python
from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial.distance import cdist
from sklearn.preprocessing import OneHotEncoder
one_hot_encoded_data = pd.get_dummies(
    data, columns=['PRODUCTLINE', 'CITY', 'COUNTRY', 'DEALSIZE'])
print(one_hot_encoded_data)
```

```python
data.drop(['PRODUCTLINE', 'CITY', 'COUNTRY', 'DEALSIZE'], axis=1, inplace=True)
```

```python
one_hot_encoded_data.drop(['QUANTITYORDERED', 'PRICEEACH',
                           'SALES', 'MONTH_ID', 'YEAR_ID'], axis=1, inplace=True)
one_hot_encoded_data.drop(['MSRP'], axis=1, inplace=True)
```

```python
data1 = data.join(one_hot_encoded_data)
```

```python
data1.head()
```

```python
distortions = []
inertias = []
mapping1 = {}
mapping2 = {}
K = range(1, 10)
```

```python
for k in K:
 # Building and fitting the model
 kmeanModel = KMeans(n_clusters=k).fit(data1)
 kmeanModel.fit(data1)

 distortions.append(sum(np.min(cdist(data1, kmeanModel.cluster_centers_,
                                     'euclidean'), axis=1)) / data1.shape[0])
 inertias.append(kmeanModel.inertia_)

 mapping1[k] = sum(np.min(cdist(data1, kmeanModel.cluster_centers_,
                                'euclidean'), axis=1)) / data1.shape[0]
 mapping2[k] = kmeanModel.inertia_
```

```python
import matplotlib.pyplot as plt
plt.plot(K, distortions, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Distortion')
plt.title('The Elbow Method using Distortion')
plt.show()
```
